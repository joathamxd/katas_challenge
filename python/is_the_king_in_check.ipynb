# Is the King in check ?

You have to write a function

> king_is_in_check()

that takes for input a 8x8 chessboard in the form of a bi-dimensional array of chars (Javascript : strings of length 1) and returns true if the black king is in check or false if it is not.

The array will include 64 squares which can contain the following characters :

- '♔' for the black King;
- '♛' for a white Queen;
- '♝' for a white Bishop;
- '♞' for a white Knight;
- '♜' for a white Rook;
- '♟' for a white Pawn;
- ' ' (a space) if there is no piece on that square.

Note : these are actually inverted-color chess Unicode characters because the dark codewars theme makes the white appear black and vice versa. Use the characters shown above.

There will always be exactly one king, which is the black king, whereas all the other pieces are white.
The board is oriented from Black's perspective.
Remember that pawns can only move and take forward.
Also be careful with the pieces' line of sight ;-) .

The input will always be valid, no need to validate it. To help you visualize the position, tests will print a chessboard to show you the problematic cases. Looking like this :

```python
def search_king_position(chessboard):
    for i, row in enumerate(chessboard):
        # for piece in row:
        if "♔" in row:
            index = row.index("♔")
            king_position = (i, index)
    return king_position
```

```python
def diag_dist(a, b):
    x = abs(a[0] - b[0])
    y = abs(a[1] - b[1])
    
    if x == y:
        return True
    return False
```

```python
def direct_dist(a, b):
    if a[0] == b[0]:
        return True
    if a[1] == b[1]:
        return True
    return False
```

```python
def queen_check(kp, qp):
    return direct_dist(kp, qp) or diag_dist(kp, qp)
```

```python
def bishop_check(kp, bp):
    return diag_dist(kp, bp)
```

```python
def knight_check(kp, knp):
    x = abs(kp[0] - knp[0])
    y = abs(kp[1] - knp[1])
    dist_knp = (x, y)
    print(dist_knp)
    valid_dist = [(1,2), (2,1)]
    if dist_knp in valid_dist: 
        return True
    return False
```

```python
def rook_check(kp, rp):
    return direct_dist(kp, rp)
```

```python
def pawn_check(kp, pp):
    if kp[0] == pp[0] + 1:
        y = abs(kp[1] - pp[1])
        return y == 1
    return False
```

```python
# Middle piece could be improve
def middle_piece(kp, op, chessboard):
    
    middle_points = []
    
    x1 = kp[0]
    y1 = kp[1]
    
    x2 = op[0]
    y2 = op[1]
    
    if x1 == x2:
        for y in range(min(y1,y2)+1, max(y1, y2)):
            middle_points.append((x1, y))
    if y1 == y2:
        for x in range(min(x1,x2)+1, max(x1, x2)):
            middle_points.append((x, y1))
            
    if diag_dist(kp, op):
        # print("Es diagonal")
        
        if x1 < x2 and y1 < y2:
            # print("Be PP")
            while x1 < x2-1 and y1 < y2-1:
                x1 = x1+1
                y1 = y1+1
                middle_points.append((x1, y1))
        elif x1 < x2 and y2 < y1:
            # print("Be PN")
            while x1 < x2-1 and y2 < y1-1:
                x1 = x1+1
                y1 = y1-1
                middle_points.append((x1, y1))
        elif x2<x1 and y2<y1:
            # print("Be NN")
            while x2 < x1-1 and y2 < y1-1:
                x2 = x2+1
                y2 = y2+1
                middle_points.append((x2, y2))
        elif x2<x1 and y1<y2:
            # print("Be NP")
            while x2 < x1-1 and y1 < y2-1:
                x2 = x2+1
                y2 = y2-1
                middle_points.append((x2, y2))

    # print(middle_points)
    
    coord_to_check = []
    
    if middle_points:
        # print(middle_points)
        for point in middle_points:
            position = chessboard[point[0]][point[1]]
            if position != " ":
                # print("Middle piece at: ", point, "Checked: " ,middle_points)
                coord_to_check.append(True)
            else:
                # print("Not middle piece for:", point)
                coord_to_check.append(False)
            
    if True in coord_to_check:
        return True
    return False
            
    # print(direcciones, "Direcciones")
    
    
```

```python
def king_is_in_check(chessboard):
    
    king_position = search_king_position(chessboard)
    # print("King Position", king_position)
    
    pieces_to_check = []
    
    for i, row in enumerate(chessboard):
        for j, piece in enumerate(row):
            piece_cord = (i, j)
            if '♟' == piece:
                pieces_to_check.append(pawn_check(king_position, piece_cord))
            elif '♞' == piece:
                pieces_to_check.append(knight_check(king_position, piece_cord))
            elif '♛' == piece:
                if not middle_piece(king_position, piece_cord, chessboard):
                    pieces_to_check.append(queen_check(king_position, piece_cord))
            elif '♝' == piece:
                if not middle_piece(king_position, piece_cord, chessboard):
                    pieces_to_check.append(bishop_check(king_position, piece_cord))
            elif '♜' == piece:
                if not middle_piece(king_position, piece_cord, chessboard):
                    pieces_to_check.append(rook_check(king_position, piece_cord))
            elif "♔" == piece or " " == piece: 
                pass
            
                    
    # print("Pieces to check: ", pieces_to_check)
    
    if True in pieces_to_check:
        return True
    return False
    
```
